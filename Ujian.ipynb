<font size="5">Neural Network</font>

```python
class Neural_Network():
    def __init__(self):
        self.input_layerSize  = 7
        self.hidden_layerSize1 = 4
        self.hidden_layerSize2 = 4
        self.output_layerSize = 1
        
        
        self.w1 = np.random.randn(self.input_layerSize,self.hidden_layerSize1)
        self.w2 = np.random.randn(self.hidden_layerSize1,self.hidden_layerSize2)
        self.w3 = np.random.randn(self.hidden_layerSize2,self.output_layerSize)
        
        #self.w1 = np.random.randn(self.input_layerSize)
        #self.w2 = np.random.randn(self.hidden_layerSize1)
        #self.w3 = np.random.randn(self.hidden_layerSize2)
        self.limit = 0.5
    
    #Set for false positve accuracy test 
        self.error_list = []
        self.true_positives = 0
        self.false_positives = 0
        self.true_negatives = 0
        self.false_negatives = 0
    
    #---------------------------------- Forward propogation
    #forward propogation
    def forward(self,X):
        self.z2 =np.dot(X, self.w1)
        self.a2 =self.sigmoid(self.z2)
        self.z3 =np.dot(self.a2,self.w2)
        self.a3 =self.sigmoid(self.z3)
        self.z4 =np.dot(self.a3,self.w3)
        yhat = self.sigmoid(self.z4)
        return yhat
    
    #sigmoid function
    def sigmoid(self,z):
        return 1/(1+np.exp(-z))
        
    #--------------------(Backpropogation) derivative of the sigmoid function and gradient of the previous sigmoid
    # Derivative Sigmoid
    def sigmoid_Prime(self, z):
        return np.exp(-z)/((1+np.exp(-z))**2)

    #Cost of passing through nodes
    def costFunction(self, X, y):
        #Cost of computation 
        self.yHat = self.forward(X)
        J = 0.5*sum((y-self.yHat)**2)
        return J
    #Cost of finishing to the other nodes  
    def costFunctionPrime(self, X, y):
        #Compute derivative with respect to W and W2 for a given X and y:
        self.yHat = self.forward(X)
        
        delta4 = np.multiply(-(y-self.yHat), self.sigmoid_Prime(self.z3))
        dJdW3 = np.dot(self.a3.T, delta4)
        
        delta3 = np.dot(delta4, self.w3)*self.sigmoid_Prime(self.z3)
        dJdW2 = np.dot(self.a2.T, delta3)
        
        delta2 = np.dot(delta3, self.w2)*self.sigmoid_Prime(self.z2)
        dJdW1 = np.dot(X.T, delta2)  
        
        return dJdW1, dJdW2, dJdW3
     
        #Helper Functions for gradient checking: -------------------------------------------------
    def getParams(self):
        #Get all weights vector:
        params = np.concatenate((self.w1.ravel(), self.w2.ravel(), self.w3.ravel()))
        return params
    
    def setParams(self, params):
        #Set W1 and W2 using single paramater vector.
        W1_start = 0
        W1_end = self.hidden_layerSize1 * self.input_layerSize
        self.W1 = np.reshape(params[W1_start:W1_end], (self.input_layerSize , self.hidden_layerSize1))
        W2_end = W1_end + self.hidden_layerSize1*self.hidden_layerSize2
        self.W2 = np.reshape(params[W1_end:W2_end], (self.hidden_layerSize1, self.hidden_layerSize2))
        W3_end = W1_end + self.hidden_layerSize2*self.output_layerSize
        self.W3 = np.reshape(params[W1_end:W3_end], (self.hidden_layerSize2, self.output_layerSize))
        
    def computeGradients(self, X, y):
        dJdW1, dJdW2, dJdW3 = self.costFunctionPrime(X, y)
        
        return np.concatenate((dJdW1.ravel(), dJdW2.ravel(),dJdW3.ravel()))
        
```

<font size="5">Checking Gradient Computation</font>

```python
def computeNumericalGradient(N, X, y):
        paramsInitial = N.getParams()
        numgrad = np.zeros(paramsInitial.shape)
        perturb = np.zeros(paramsInitial.shape)
        e = 1e-4

        for p in range(len(paramsInitial)):
            #Set perturbation vector
            perturb[p] = e
            N.setParams(paramsInitial + perturb)
            loss2 = N.costFunction(X, y)
            
            N.setParams(paramsInitial - perturb)
            loss1 = N.costFunction(X, y)

            #Compute Numerical Gradient
            numgrad[p] = (loss2 - loss1) / (2*e)

            #Return the value we changed to zero:
            perturb[p] = 0
            
        #Return Params to original value:
        N.setParams(paramsInitial)

        return numgrad 
```

<font size="5">Training</font>

```python
from scipy import optimize
```

```python
class trainer(object):
    def __init__(self, N):
        #Make Local reference to network:
        self.N = N
        
    def callbackF(self, params):
        self.N.setParams(params)
        self.J.append(self.N.costFunction(self.X, self.y))   
        
    def costFunctionWrapper(self, params, X, y):
        self.N.setParams(params)
        cost = self.N.costFunction(X, y)
        grad = self.N.computeGradients(X,y)
        
        return cost, grad
        
    def train(self, X, y):
        #Make an internal variable for the callback function:
        self.X = X
        self.y = y

        #Make empty list to store costs:
        self.J = []
        
        params0 = self.N.getParams()

        options = {'maxiter': 200, 'disp' : True}
        _res = optimize.minimize(self.costFunctionWrapper, params0, jac=True, method='BFGS', \
                                 args=(X, y), options=options, callback=self.callbackF)

        self.N.setParams(_res.x)
        self.optimizationResults = _res
```

```python
T = trainer(NN)
```
